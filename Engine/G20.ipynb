```python
import pandas as pd
from sklearn.ensemble import IsolationForest

# Assuming 'df' is your cleaned DataFrame with numeric columns

# Select only numeric columns
numeric_columns = df.select_dtypes(include='number').columns

# Create a DataFrame containing only numeric columns
numeric_df = df[numeric_columns]

# Initialize the Isolation Forest model
contamination_factor = 0.1  # Set your contamination factor
model = IsolationForest(contamination=contamination_factor, random_state=42)

# Fit the model and predict outliers
outliers = model.fit_predict(numeric_df)

# Create a mask for non-outliers
non_outliers_mask = outliers != -1

# Filter the original DataFrame to keep only non-outliers for numeric columns
df_no_outliers = df[non_outliers_mask]

# Display the resulting DataFrame without outliers for numeric columns
print(df_no_outliers)
```

```python
import pandas as pd
from sklearn.ensemble import IsolationForest

# Assuming 'df' is your cleaned DataFrame with numeric columns

# Select only numeric columns
numeric_columns = df.select_dtypes(include='number').columns

# Create a DataFrame containing only numeric columns
numeric_df = df[numeric_columns]

# Transpose the DataFrame so that each column is treated as an instance
numeric_df_transposed = numeric_df.T

# Initialize the Isolation Forest model
contamination_factor = 0.1  # Set your contamination factor
model = IsolationForest(contamination=contamination_factor, random_state=42)

# Fit the model and predict outliers for columns
outliers_columns = model.fit_predict(numeric_df_transposed)

# Create a mask for non-outlier columns
non_outliers_columns_mask = outliers_columns != -1

# Filter the original DataFrame to keep only non-outlier columns for numeric columns
df_no_outliers_columns = df[numeric_columns].loc[:, non_outliers_columns_mask]

# Display the resulting DataFrame without outlier columns for numeric columns
print(df_no_outliers_columns)
```
